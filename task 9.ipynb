### Importing Libraries and Loading Data

```python
import pandas as pd
data = pd.read_csv('Smartphones_cleaned_dataset.csv')
```

```python

shape = data.shape

columns = data.columns

describe = data.describe()

null_values = data.isnull().sum()

data_types = data.dtypes

catagorical_columns = data.select_dtypes(include=['object']).columns

numerical_columns = data.select_dtypes(exclude=['object']).columns

print("Shape of the dataset: \n", shape)
print("-------------------------------")
print("Columns in the dataset: \n", columns)
print("-------------------------------")
print("description of the dataset: \n", describe)
print("-------------------------------")
print("Null values in each column: \n", null_values)
print("-------------------------------")
print("Data types of each column:\n", data_types)
print("-------------------------------")
print("Categorical columns: \n", catagorical_columns)
print("-------------------------------")
print("Numerical columns: \n", numerical_columns)
```
