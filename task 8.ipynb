```python
import math
values = [3,2,5,6,7,7,8,12]
lenght = len(values)
levels = math.ceil(math.log2(lenght))
print("levels:",levels)

def minmax(current_level,index,minturns,value,total_levels):
    if current_level == total_levels:
        return values[index]
    if minturns:
        return min(minmax(current_level+1,index*2,False,values,total_levels),
                     minmax(current_level+1,index*2+1,False,values,total_levels))
    else:
        return max(minmax(current_level+1,index*2,True,values,total_levels),
                     minmax(current_level+1,index*2+1,True,values,total_levels))
    
print(f"The value is: {minmax(0,0,True,values,levels)}")

       
```
