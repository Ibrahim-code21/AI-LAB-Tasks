```python
class Movie:
    def __init__(self, movies):
        self.movies = movies

    def add_movies(self):
        # it  asks if u wanna add a movie or not
        while True:
            choice = input("Would you like to add some movies in this dataset (y/n): ")
            if choice.lower() == 'y':
                title = input("Enter the movie title: ")
                budget = int(input("Enter the movie budget: "))
                self.movies.append((title, budget))
                print(f"{title} with a budget of ${budget} has been added.")
                print(self.movies)
            elif choice.lower() == 'n':
                break
            else:
                print("Invalid input. Please enter 'y' or 'n'.")

    def average(self):
        # This caculate average
        total_budget = sum(budget for _, budget in self.movies)
        return total_budget / len(self.movies)

    def above_average(self):
        # This first caculate average then add see which movies are above average
        average = self.average()
        print("The total average budget for all movies is:")
        print("-------------------------------------------------")
        print(f"Average movie budget: ${average}")
        
        print("-------------------------------------------------")
        print("These movies had above average budgets:")
        for title, budget in self.movies:
            if budget > average:
                print(f"{title} had an above average budget of ${budget}")
                print(f"which is ${budget - average} above the average")
        
        print("-------------------------------------------------")
        print("These are the titles with above average budgets:")     
        for title, budget in self.movies:
            if budget > average:
                print(title)
        print("-------------------------------------------------")
movies = [ 
    ("Eternal Sunshine of the Spotless Mind", 20000000), 
    ("Memento", 9000000), 
    ("Requiem for a Dream", 4500000), 
    ("Pirates of the Caribbean: On Stranger Tides", 379000000), 
    ("Avengers: Age of Ultron", 365000000), 
    ("Avengers: Endgame", 356000000), 
    ("Incredibles 2", 200000000) 
]
movie_1 = Movie(movies)
print("enter what you want to do: add/display/average:")
print("1. Add movies\n2. Calculate average budget\n3. Display movies with above average budgets\n4. Exit")

# this is a menu to select your desired operation
menu = input("enter what you want to do: 1/2/3/4: ")
while True:
    menu = input("enter what you want to do: 1/2/3/4: ")
    match menu:
        case "1":
            movie_1.add_movies()
        case "2":
            print("Average budget is:", movie_1.average())
        case "3":   
            movie_1.above_average()
        case "4":
            print("Exiting the program.")
            break
        case _:
            print("Invalid input. please enter 1, 2, 3 or 4")




```
